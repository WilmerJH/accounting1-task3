```python
import pandas as pd
from pathlib import Path

# check root menu
root = Path.cwd()
print(root)
```

```python
# read in edgar csv

csv_path = Path("../../data/external/10k_word_counts.csv")

df = pd.read_csv(
    csv_path,
    dtype={"cik": str}      # make sure cik would not be read as integer
)
df.head()
```

```python
# ensure cik to be 10-digit

df["cik"] = (
    df["cik"]
    .str.strip()
    .str.replace(r"\.0$", "", regex=True)
    .str.zfill(10)
)

df[["cik", "name"]].head()
```

```python
# extract unique cik number

cik_list = (
    df["cik"]
    .dropna()
    .drop_duplicates()
    .sort_values()
    .reset_index(drop=True)
)

print("Number of unique CIKs:", len(cik_list))
cik_list.head()
```

```python
# export cik number to csv

out_dir = Path("../../data/pulled")
out_dir.mkdir(exist_ok=True)

cik_list.to_frame(name="cik").to_csv(out_dir / "cik_list_all.csv", index=False)

with open(out_dir / "cik_list.txt", "w", encoding="utf-8") as f:
    for cik in cik_list:
        f.write(cik + "\n")
```
